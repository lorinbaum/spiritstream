## Print readable ttf programs

```python
from ttf import Interpreter
II =  Interpreter(fontfile="assets/fonts/Fira_Code_v6.2/ttf/FiraCode-Regular.ttf")
II.fontsize = 12
II.dpi = 144
II.run_program(II.tables["fpgm"], "fpgm")
II.run_program(II.tables["prep"], "prep")
```

```python
II.debug = True
II.getGlyph(0x44)
```

```python
pgm = II.fpgm
# pgm = II.prep
# pgm = II.glyf



def next(i):
    i += 1
    return i, pgm[i]
def pInst(x): print(f"{'  ' * depth + str(x)}", end=" ")

i = -1
depth = 0
while i < len(pgm):
    i, op = next(i)
    name = II.opnames[op]
    if name in ["ENDF", "ELSE", "EIF"]: depth -= 1
    pInst(name)
    if name == "NPUSHB":
        i, n  = next(i)
        for _ in range(n):
            i, v = next(i)
            print(hex(v), end=" ")
        print("")
    elif name == "NPUSHW":
        i, n  = next(i)
        for _ in range(0, n*2, 2):
            i, v1 = next(i)
            i, v2 = next(i)
            v = (i << 8) | v2
            print(hex(v), end=" ")
        print("")
    elif name.startswith("PUSHB"):
        n = int(name[-3]) + 1
        for _ in range(n):
            i, v = next(i)
            print(hex(v), end=" ")
        print("")
    elif name.startswith("PUSHW"):
        n = int(name[-3]) + 1
        for _ in range(0, n*2, 2):
            i, v1 = next(i)
            i, v2 = next(i)
            v = (v1 << 8) | v2
            print(hex(v), end=" ")
        print("")
    else: print("")
    if name in ["FDEF", "IF", "ELSE"]: depth += 1
```

## P5.js script to display a glyph

```python
def scaleX(x): return round((x / maxScale) * height + height)
def scaleY(y): return round(((height - y) / maxScale) * height + height)
height = 400 // 2
maxScale = 1500
contours = [-1] + I.g.endPtsContours
print("function setup() {")
print(f"createCanvas({height*2}, {height*2})")
print("background(0)")
print("stroke(255)")
print("strokeWeight(5)")
for start, end in zip(contours, contours[1:]):
    prevX = prevY = 0
    x1 = y1 = 0
    for i in range(start+1, end+1):
        p = I.g.get_point(i)
        p.x, p.y = scaleX(p.x), scaleY(p.y)
        print(f"point({p.x},{p.y})")
        if i == start+1: x1, y1 = p.x, p.y
        else: print(f"line({prevX}, {prevY}, {p.x}, {p.y})")
        prevX, prevY = p.x, p.y
    print(f"line({prevX}, {prevY}, {x1}, {y1})")
    print("// ----------------------------")
print("stroke(255, 0, 0)")
print("strokeWeight(10)")
ph1 = I.g.get_point(contours[-1] + 1)
ph2 = I.g.get_point(contours[-1] + 2)
ph1.x = scaleX(ph1.x)
ph1.y = scaleY(ph1.y)
ph2.x = scaleX(ph2.x)
ph2.y = scaleY(ph2.y)
print(f"point({ph1.x}, {ph1.y})")
print(f"point({ph2.x}, {ph2.y})")
print("}")
```
